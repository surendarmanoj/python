```python
n = int(input("Enter the number of elements you are planning to process : "))
val = []
for i in range(n):
    x = input("Enter the value : ")
    val.append(x)
print(val)
def no_string(lst):
    op_lst = []
    for i in lst:
        if i.isnumeric():
            op_lst.append(i)
    return op_lst
print(no_string(val))
```

```python
string = input("Enter the string to be reversed : ")
def reverse(ip):
    str1 = ""
    for i in range(len(ip)):
        str1 = str1 + ip[len(ip)-1-i]
    return str1
print(reverse(string))
```

```python
val = int(input("Enter the number : "))
def factorial(n):
    if n == 1:
        return 1
    else:
        return n*factorial(n-1)
print(factorial(val))
```

```python
n = int(input("Enter the number of elements you are planning to process : "))
val = []
for i in range(n):
    x = input("Enter the value : ")
    val.append(x)
print(val)

type1 = input("Enter the element to be grouped")

def last(lst, tpe):
    op = []
    counter = 0
    for i in lst:
        
        if i == tpe:
            counter = counter +1
        else:
            op.append(i)
    for i in range(counter):
        op.append(tpe)
    return op

print(last(val, type1))
```
