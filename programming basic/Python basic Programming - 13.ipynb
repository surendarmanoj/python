```python
x = int(input('Enter the number of rows'))
y = int(input('Enter the number of columns'))

z = []
counter = 0
for i in range(x):
    z1 = []
    
    for j in range(y):
        z1.append(j*counter)
    z.append(z1)
    counter += 1

print(z)
```

```python
x = str(input("Enter the string : "))
num_cnt = 0
str_cnt = 0

for i in x:
    if i.isalpha():
        str_cnt += 1
    elif i.isnumeric():
        num_cnt += 1
        
print(num_cnt)
print(str_cnt)
```

```python
import re

x = input("Enter the passwords in comma seperated format : ")
y = x.split(",")

accepted_pass = []
for i in y:
    
    if len(i) < 6 or len(i) > 12:
        continue

    elif not re.search("([a-z])+", i):
        continue

    elif not re.search("([A-Z])+", i):
        continue

    elif not re.search("([0-9])+", i):
        continue

    elif not re.search("([!@$%^&])+", i):
        continue

    else:
        accepted_pass.append(i)

print((" ").join(accepted_pass))
        
```
