```python
n = int(input('Enter the number range : '))
def divChecker(n):
    for i in range(n):
        if i % 7 == 0:
            yield i


for j in divChecker(n):
    print(j)
```

```python
text = input("Enter the string to be processed : ")
x = text.split(" ")

op = []
for i in x:
    count = 0
    for j in x:
        if i==j:
            count = count+1
    y = op.count(str(i)+':'+str(count))
    if y == 0:
        op.append(str(i)+':'+str(count))

print(op)
```

```python
class Person(object):
    def getGender( self ):
        return "Unknown"

class Male( Person ):
    def getGender( self ):
        return "Male"

class Female( Person ):
    def getGender( self ):
        return "Female"

aMale = Male()
aFemale= Female()
print (aMale.getGender())
print (aFemale.getGender())
```

```python
subjects=["I", "You"]
verbs=["Play", "Love"]
objects=["Hockey","Football"]

res = [[i, j, k] for i in subjects  
                 for j in verbs 
                 for k in objects] 
for x in res:
    print(" ".join(x))
```

```python
from bisect import bisect_left
  
def BinarySearch(a, x):
    i = bisect_left(a, x)
    if i != len(a) and a[i] == x:
        return i
    else:
        return -1
  
a  = [1, 2, 4, 4, 8]
x = int(4)
res = BinarySearch(a, x)
if res == -1:
    print(x, "is absent")
else:
    print("First occurrence of", x, "is present at", res)
```
