```python


def splitArr(arr, n, pos):
    for i in range(0, pos):
        x = arr[0]
        for j in range(0, n-1):
            arr[j] = arr[j + 1]
        
        arr[n-1] = x


# main
arr = [12, 10, 5, 6, 52, 36]
n = len(arr)
position = 2

splitArr(arr, n, position)

for i in range(0, n):
    print(arr[i], end = ' ')


```

```python

def isMonotonic(A):

    return (all(A[i] <= A[i + 1] for i in range(len(A) - 1)) or
            all(A[i] >= A[i + 1] for i in range(len(A) - 1)))

A = [6, 5, 4, 7]

print(isMonotonic(A))
```
