```python
def cnt(a,b,c):
    if a == b == c:
        return 3
    elif (a==b) or b==c or c == a:
        return 2
    else:
        return 0
cnt(2,2,2)
```

```python
dict1 = { "D": 1,"B": 2,"C": 3}
 
def dict_to_tuple(d):
    list1 = [(k, v) for k, v in d.items()]
    return list1
 
print(dict_to_tuple(dict1))
```

```python
def mapping(val):
    f = {}
    for i in val:
        f[i] = i.upper()
    return f
mapping(['a','b','c'])
```

```python
vowels = ['a','e','i','o','u']
def replace_vowels(string, r_v):
    op = ""
    for i in string:
        if vowels.count(i) != 0:
            op = op + r_v
        else:
            op = op + i    
    return op
replace_vowels("apples and bananas", "u")
```

```python
def ascii_capitalize(string):
    op = ""
    for i in string:
        if ord(i)%2 == 0:
            op = op + i.upper()
        else:
            op = op + i
    return op
ascii_capitalize("Surendar Manoj")
```
