```python
data = input("Enter the string to be processed : ")
def duplicate_char(string):
    op = ""
    for i in string:
        for j in range(2):
            op = op+i
    return op
duplicate_char(data)
```

```python
def is_bool(data):
    if type(data) == type(True):
        if(data == True):
            return False
        else:
            return True
    else:
        return "boolean expected"

is_bool("a")
        
```

```python
def fold(n):
    return (.5*(2**n))/1000
fold(21)
```

```python
ip = input("Enter the value")

def index_upper(data):
    indexes = []
    for i in range(len(data)):
        if data[i].isupper():
            indexes.append(i)
    return indexes

index_upper(ip)
```
