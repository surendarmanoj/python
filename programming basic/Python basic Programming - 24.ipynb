```python
def amplify(n):
    x = []
    for i in range(n):
        if i%4 == 0:
            x.append(i*10)
        else:
            x.append(i)
    return x
amplify(10)
```

```python
def unique(val):
    x = []
    for i in val:
        if val.count(i) == 1:
            x.append(i)
    return x
for i in unique([3, 3, 3, 7, 3, 3]):
    print(i)
```

```python
rad = float(input("Enter the Radius of the circle : "))
class circle:
    def __init__(self,radius):
        self.radius = radius
    def area(self):
        return (22*self.radius**2)/7
    def perimeter(self):
        return (22*2*self.radius)/7
c = circle(rad)
print(c.area())
print(c.perimeter())
```

```python
def Sorting(lst):
    lst2 = sorted(lst, key=len)
    return lst2
      
lst = ['Google','Apple','Microsoft']
print(Sorting(lst))
```

```python
def pithogaras(x,y,z):
    l = [x,y,z]
    l.sort()
    op = (l[0]**2+l[1]**2)
    if op == l[2]**2:
        return True
    else:
        return False

pithogaras(3,5,4)
```
