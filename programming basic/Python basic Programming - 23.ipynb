```python
x = input("Enter the number to be checked : ")
def reverse(string):
    temp = ""
    for i in range(len(string)):
        temp = temp + string[len(string)-1-i]
    if temp == string:
        return True
    else:
        return False
reverse(x)
```

```python
def prod(val):
    val1 = val.split(", ")
    x = 1
    for i in val1:
        x = x*int(i)
    return x
prod("1, 2, 3, 4")
```

```python
def original(val):
    x = list(set(val))
    return x
original([1, 3, 3, 5, 5])
```

```python
def mean(n):
    x = 0
    for i in n:
        x = x+int(i)
    y = x/len(n)
    return y
ip = input("Enter the number : ")
mean(ip)
```
