```python
def fix_list(lst, x):
    del lst[0]
    lst.append(x)
    return lst

fix_list([1,2,3,4,5],0)
    
```

```python
def budget(lst):
    bdgt = 0
    for i in lst:
        bdgt = bdgt + i.get('budget')
    return bdgt
budget([
    { 'name': 'John', 'age': 21, 'budget': 23000 },
    { 'name': 'Steve', 'age': 32, 'budget': 40000 },
    { 'name': 'Martin', 'age': 16, 'budget': 2700 }])
```

```python
def arrange(string):
    x = [i for i in string]
    x.sort()
    str1 = ""
    for i in x:
        str1 = str1 + i
    return str1

print(arrange('surendarmanoj'))
```

```python
def compound_int(p,t,r,n):
    return(p*((n+r)/n)**(n*t))
print(compound_int(10000,10,.06,12))
```

```python
n = int(input("Enter the number of elements you are planning to process : "))
val = []
for i in range(n):
    x = input("Enter the value : ")
    val.append(x)
print(val)
def no_string(lst):
    op_lst = []
    for i in lst:
        if i.isnumeric():
            op_lst.append(i)
    return op_lst
print(no_string(val))
```
