```python
def div_range(a,b,c):
    x = 0
    for i in range(a,b):
        if i%c == 0:
            x = x+i
    return x

div_range(1,20,2)
```

```python
def correct_signs(expression):
    if(expression):
        return True
    else:
        return False
    
correct_signs(13 > 44 > 33 > 1)
```

```python
vowels = ['a','e','i','o','u']
def replace_vowels(string):
    op_str = ""
    for i in string:
        if (i in vowels):
            op_str = op_str + '#'
        else:
            op_str = op_str + i
    return op_str

replace_vowels("Surendar Manoj")
```

```python
def factorial(n):
    if n == 1:
        return n
    else:
        return n * factorial(n-1)
    
x = int(input("Enter a positive number"))
if x > 0:
    print(factorial(x))
```

```python
a = input("Enter the string A : ")
b = input("Enter the string B : ")
def hamming(a1,b1):
    counter = 0
    for i in range(len(a1)):
        
        if a1[i]!=b1[i]:
            counter = counter +1
    return counter

print("The hamming distance between the strings is {}".format(hamming(a,b)))
```
