```python
def checkIfCurzonNumber(N):
 
    powerTerm, productTerm = 0, 0
    powerTerm = pow(2, N) + 1
    productTerm = 2 * N + 1
 
    if (powerTerm % productTerm == 0):
        print("Yes")
    else:
        print("No")
 
checkIfCurzonNumber(44)
```

```python
def area_of_hexagon(side):
    return(3**1.5*side**2)/2
area_of_hexagon(1)
```

```python
bin_value = input("Enter the binary value : ")
def bin_to_dec(binary):
    x = 0
    counter = 0
    for i in binary:
        x = int(x) + int((int(i)*(2**counter)))
        counter = counter + 1
    return x
        
bin_to_dec(bin_value)
```
