```python
def div_range(x,y,z):
    lst = []
    for i in range(x,y):
        if i%z == 0:
            lst.append(i)
    return lst
div_range(1,10,3)
```

```python
def sim(lst, lst1):
    count = 0
    for i in lst:
        if lst1.count(i) == 0:
            count += 1
    if count == 1:
        return True
    else:
        return False

sim([1, 2, 3, 4, 5], [0, 5, 1, 2, 3])
```

```python
def sec_soc(val):
    str1 = ""
    for i in val:
        str1 = str1 + i[0]
    return str1
sec_soc(["Adam","Bravo","Charlie"])
```

```python
def chk(string):
    x = list(string)
    if len(x) == len(set(x)):
        return True
    else:
        return False
chk("Oscar")
```

```python
def order(string):
    x = list(string)
    x.sort()
    op = ""
    for i in x:
        op = op + i
    if (string == op):
        return True
    else:
        return False
order('asd')
```
