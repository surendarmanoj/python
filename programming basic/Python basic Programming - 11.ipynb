```python
def string_k(k, str):
    string = []
    text = str.split(" ")
    for x in text:
        if len(x) > k:
            string.append(x)
    return string
 
k = 3
string ="Hi.. I am Surendar Manoj"
print(string_k(k, string))
```

```python
def remove(string, i): 
    a = string[ : i] 
    b = string[i + 1: ]
    return a + b
if __name__ == '__main__':
    string = "Hi.. I am Surendar Manoj"
    i = 5
    print(remove(string, i))
```

```python
def split_string(string):
    list_string = string.split(' ')
    return list_string

def join_string(list_string):
    string = '-'.join(list_string)
    return string

if __name__ == '__main__':
    string = 'Hi.. I am Surendar Manoj'
    list_string = split_string(string)
    print(list_string)

    new_string = join_string(list_string)
    print(new_string)
```

```python
def UncommonWords(A, B):
    count = {}
    for word in A.split():
        count[word] = count.get(word, 0) + 1
      
    for word in B.split():
        count[word] = count.get(word, 0) + 1
  
    return [word for word in count if count[word] == 1]
  
A = "I am Surendar Manoj"
B = "I am a python Big Data Developer"
  
print(UncommonWords(A, B))
```
