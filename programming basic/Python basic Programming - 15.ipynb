```python
n = int(input('Enter the number range : '))
def divChecker(n):
    for i in range(n):
        if i % 7 == 0 and i%5==0:
            yield i


for j in divChecker(n):
    print(j)
```

```python
def EvenGenerator(n):
    i=0
    while i<=n:
        if i%2==0:
            yield i
        i+=1


n=int(input("Enter the number : "))
values = []
for i in EvenGenerator(n):
    values.append(str(i))

print (",".join(values))
```

```python
class Shape(object):
    def __init__(self):
        pass

    def area(self):
        return 0

class Square(Shape):
    def __init__(self, l):
        self.length = l

    def area(self):
        return self.length*self.length

aSquare= Square(12)
print (aSquare.area())
```
