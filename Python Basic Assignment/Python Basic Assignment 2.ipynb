11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

range(10) -> Produces the output from 0 to the value defined within the range
range(0,10) -> produces the output from starting number(first parameter) to the ending number(next parameter)
range(0,10,1) -> Produces output from starting number to ending number with the defined step(Third Parameter.)
