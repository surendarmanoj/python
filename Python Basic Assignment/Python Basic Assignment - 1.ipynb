3. Describe three different data types.

Number Data Types:
    There are three types of numbers - Complex numbers, floating datatypes and integers
    Complex numbers have a imaginary part and a real part.
    Floating numbers are the ones with decimal fields.It is denoted as float
    Integers are the representation of whole numbers. It is denoted as int.
    
    Each of these number tyrpes have their own significance based on the problem 
    in which they are used and the output expected.
    
String Data Types:
    String is a sequence of characters. It is denoted by str.
    It is commonly enclosed between "" or ''. 
    All characters enclosed between the quotes are part of string including spaces and special characters.
    There is no limitation on the number of characters in the string. 
    The only limitation will be the memory resource of the machine processing the data

Boolean Data Types:
    They can accept only two values - True or False
    These Data types are used to make decisions while the program is getting executed.
    They are denoted by bool.
    They are the outcome of most of the logical operators and condition statements.
    They are used to make the flow of the program smart.
