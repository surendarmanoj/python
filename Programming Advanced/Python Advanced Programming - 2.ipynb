```python
def pentagonal(n):
    return ((5*n*n) -(5*n) +2)/2

pentagonal(8)
```

```python
def encrypt(x):
    y = ""
    z = len(x)
    vow = {'a':0,'e':1,'i':2,'o':3,'u':4}
    for i in range(z):
        a = vow.get(x[i])
        
        if a == None:
            y = y+x[z-1]
        else:
            y = y+str(a)
        z = z-1
            
    print(y+"aca")
    
encrypt('apple')
```

```python
import datetime

def has_friday_13(month, year):
    return datetime.date(year, month, 13).strftime("%A") == "Friday"
has_friday_13(3,2020)
```

```python
import re
lst = ["bad cookie", "good cookie", "bad cookie", "good cookie", "good cookie"]
pattern = r'(?<!good)\scookie'# Regex Negative lookbehind expression
data = re.findall(pattern,' '.join(lst))
print(f'No of Bad cookies produced per day ➞ {len(data)}')
```

```python
def pluralise(x):
    y = set()
    for i in x:
        if x.count(i)>1:
            y.add(i+"s")
        else:
            y.add(i)
    return y
pluralise(["cow", "pig", "cow", "cow"])
```
